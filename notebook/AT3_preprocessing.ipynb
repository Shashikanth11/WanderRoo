## Reviews Dataset: Preprocessing and Data Preperation

```python
review_df.head()
```

```python
review_df.info()
```

```python
important_columns = ['listing_id', 'date', 'comments']
clean_review_df = review_df[important_columns]
```

```python
clean_review_df.head()
```

```python
clean_review_df.info()
```

```python
# Convert date column to datetime
clean_review_df['date'] = pd.to_datetime(clean_review_df['date'])

# Remove rows with NaN comments
clean_review_df = clean_review_df.dropna(subset=['comments'])

# Sort by listing_id and date descending (latest first)
clean_review_df = clean_review_df.sort_values(by=['listing_id', 'date'], ascending=[True, False])

# Group by listing_id and take the latest 15 reviews
latest_reviews = clean_review_df.groupby('listing_id').head(15)

# Combine reviews per listing_id
grouped_reviews = latest_reviews.groupby('listing_id')['comments'].apply(lambda x: "\n".join(x)).reset_index()

# View full content
pd.set_option("display.max_rows", None)
pd.set_option("display.max_colwidth", None)

```
